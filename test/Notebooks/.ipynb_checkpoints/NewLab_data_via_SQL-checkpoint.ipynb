### SQL setup
create engine for CBAS db

```python
passwd = sqlconfig.passwd  # From sqlconfig.py
user = sqlconfig.user  # From sqlconfig.py
DB = 'CBAS'  #name of databases to activate 
```

```python
engine = sqlalchemy.create_engine('mysql+pymysql://'+user+':'+passwd+'@34.67.192.90/'+DB)
```

```python
# SQL syntax select ALL data in table
query= ''' 
SELECT * 
FROM NewLab
'''
```

```python
#place query in CBAStest df

CBAStest =  pd.read_sql(query,engine,
                        parse_dates=["timestamp"],
                        index_col=["timestamp"])
```
